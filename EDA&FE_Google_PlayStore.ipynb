```python
df_copy['Reviews']=df_copy['Reviews'].astype('int')
```

```python
# 19M= 19000
```

```python
df_copy['Size']=df_copy['Size'].str.replace('M','1000')
df_copy['Size']=df_copy['Size'].str.replace('k','')
df_copy['Size']=df_copy['Size'].replace('Varies with device',np.nan)
df_copy['Size']=df_copy['Size'].astype('float')
```

```python
df_copy.info()
```

```python
df_copy['Installs'].unique()  # this is not a cline data, here + and $ character are remove
```

```python
df_copy['Price'].unique()
```

```python
chars_to_remove=['+',',','$']
cols_to_clean=['Installs','Price']
for item in chars_to_remove:
  for col in cols_to_clean:
    df_copy[col]=df_copy[col].str.replace(item,'')
```

```python
df_copy['Price'].unique()
```

```python
df_copy['Installs'].unique()
```

```python
df_copy['Installs']=df_copy['Installs'].astype('int')  # ti is convert in to the interger data type
```

```python
df_copy['Price']=df_copy['Price'].astype('float')
```

```python
df_copy.info()
```

```python
#change the date Day, months, year

df_copy['Last Updates']=pd.to_datetime(df_copy['Last Updated'])
df_copy['Day']=df_copy['Last Updates'].dt.day
df_copy['Month']=df_copy['Last Updates'].dt.month
df_copy['Year']=df_copy['Last Updates'].dt.year
```

```python
df_copy=df_copy.drop_duplicates(subset=['App'],keep='first')  # remove the duplicate value
```

```python
## Lets go ahead and explore more data
numeric_features=[feature for feature in df_copy.columns if df_copy[feature].dtype!='O']
categorical_features=[feature for feature in df_copy.columns if df_copy[feature].dtype=='O']

# print columns
print('We have {} numerical feature : {}'.format(len(numeric_features),numeric_features))
print('We have {} categorical feature : {}'.format(len(categorical_features),categorical_features))
```

```python
## Visualization Diagrams

plt.figure(figsize=(15,15))
plt.suptitle('Univeriate Analysis of Numerical features', fontsize=20,fontweight='bold',alpha=0.8,y=1.)

for i in range(0, len(numeric_features)):
  plt.subplot(3,3,i+1)
  sns.histplot(df_copy[numeric_features[i]],kde=True)
  plt.xlabel(numeric_features[i])
  plt.tight_layout()
```

##Observation

Rating and Year is left skewes while Reviews, Size, installs and Price are right skewed

```python
## Visualization Diagrams

plt.figure(figsize=(15, 15))
plt.suptitle('Univariate Analysis of Numerical Features', fontsize=20, fontweight='bold', alpha=0.8, y=1.)

for i in range(0, len(numeric_features)):
    plt.subplot(5, 3, i+1)
    sns.kdeplot(x=df_copy[numeric_features[i]],shade=True, color='r')
    plt.xlabel(numeric_features[i])
    plt.tight_layout()
```

## Which Category has largest number of installations??

```python
df_cat_installs = df_copy.groupby(['Category'])['Installs'].sum().sort_values(ascending = False).reset_index()
df_cat_installs.Installs = df_cat_installs.Installs/1000000000# converting into billions
df2 = df_cat_installs.head(10)
plt.figure(figsize = (14,10))
sns.set_context("talk")
sns.set_style("darkgrid")
ax = sns.barplot(x = 'Installs' , y = 'Category' , data = df2 )
ax.set_xlabel('No. of Installations in Billions')
ax.set_ylabel('')
ax.set_title("Most Popular Categories in Play Store", size = 20)
```
